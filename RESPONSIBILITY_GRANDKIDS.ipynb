```python
def read_file(file, year):
    df = pd.read_csv(file)
    df = df.iloc[:2]
    df['Label'] = df['Label'].apply(lambda x: x.lstrip())
    df['Estimate'] = df['Estimate'].apply(lambda x: x.rstrip('%'))
    df['Estimate'] = df['Estimate'].apply(lambda x: x.replace(',', ''))
    df['Estimate'] = df['Estimate'].astype(float)    
    df.drop(['Margin of Error'], axis=1, inplace=True)
    df.set_index('Label', inplace=True)
    df = df.T
    df.index = [year]
    try:
        df.rename(columns={'Living with grandchild(ren)': 'Grandparents living with grandchild(ren)'}, inplace=True)
    except:
        pass
    df.rename(columns={'Grandparents living with grandchild(ren)': 'Grandparents Living with Grandchild(ren)'}, inplace=True)

    return df
```

```python
read_file('data/2010/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2010)
```

```python
for year in range(2010, 2023):
    file = 'data/'+str(year)+'/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV'
    
    try:
        read_file(file, year)
    except:
        print(f"No file for year '{year}'")
```

```python
RESPONSIBILITY_FOR_GRANDKIDS_2010 = read_file('data/2010/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2010)
RESPONSIBILITY_FOR_GRANDKIDS_2011 = read_file('data/2011/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2011)
RESPONSIBILITY_FOR_GRANDKIDS_2012 = read_file('data/2012/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2012)
RESPONSIBILITY_FOR_GRANDKIDS_2013 = read_file('data/2013/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2013)
RESPONSIBILITY_FOR_GRANDKIDS_2014 = read_file('data/2014/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2014)
RESPONSIBILITY_FOR_GRANDKIDS_2015 = read_file('data/2015/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2015)
RESPONSIBILITY_FOR_GRANDKIDS_2016 = read_file('data/2016/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2016)
RESPONSIBILITY_FOR_GRANDKIDS_2017 = read_file('data/2017/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2017)
RESPONSIBILITY_FOR_GRANDKIDS_2018= read_file('data/2018/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2018)
RESPONSIBILITY_FOR_GRANDKIDS_2019= read_file('data/2019/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2019)
RESPONSIBILITY_FOR_GRANDKIDS_2021= read_file('data/2021/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2021)
RESPONSIBILITY_FOR_GRANDKIDS_2022= read_file('data/2022/RESPONSIBILITY_FOR_GRANDCHILDREN_UNDER_18_YEARS.CSV', 2022)
```
