```python
def read_file(file, year):
    df = pd.read_csv(file)
    df['Label'] = df['Label'].apply(lambda x: x.lstrip())
    df['Estimate'] = df['Estimate'].apply(lambda x: x.rstrip('%'))
    df['Estimate'] = df['Estimate'].apply(lambda x: x.replace(',', ''))
    df['Estimate'] = df['Estimate'].astype(float)
    df.drop(columns=['Margin of Error'], inplace=True)
    df.set_index('Label', inplace=True)
    df = df.T
    return df
```

```python
for year in range(2010, 2023):
    file = 'data/'+str(year)+'/YEAR_STRUCTURE_BUILT.CSV'
    
    try:
        read_file(file, year)
    except:
        print(f"No file for year '{year}'")
```
