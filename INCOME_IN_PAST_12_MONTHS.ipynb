```python
def income_past_12_months(file, year):
    df = pd.read_csv(file)
    df['Label'] = df['Label'].apply(lambda x: x.lstrip())
    df_1 = df.iloc[:16]
    df_1['Estimate'] = df_1['Estimate'].apply(lambda x: x.rstrip('%'))
    df_1['Estimate'] = df_1['Estimate'].apply(lambda x: x.replace(',', ''))
    df_1['Estimate'] = df_1['Estimate'].astype(float)
    df_1.drop(columns=['Margin of Error'], inplace=True)
    df_1.set_index('Label', inplace=True)
    df_2 = df_1.T
    df_2.index = [year]
    return df_2
```

```python
for year in range(2010, 2023):
    file = 'data/' + str(year) + '/INCOME_IN_THE_PAST_12_MONTHS_IN_' + str(year) + '_INFLATIONADJUSTED_DOLLARS.CSV'
    
    try: 
        income_past_12_months(file, year)
    except:
        print(f"No file for year '{year}'")
```

```python
income_past_12_months('data/2010/INCOME_IN_THE_PAST_12_MONTHS_IN_2010_INFLATIONADJUSTED_DOLLARS.CSV', 2010)
```

```python
INCOME_PAST_2022 = income_past_12_months('data/2022/INCOME_IN_THE_PAST_12_MONTHS_IN_2022_INFLATIONADJUSTED_DOLLARS.CSV', 2022)
INCOME_PAST_2021 = income_past_12_months('data/2021/INCOME_IN_THE_PAST_12_MONTHS_IN_2021_INFLATIONADJUSTED_DOLLARS.CSV', 2021)
INCOME_PAST_2019 = income_past_12_months('data/2019/INCOME_IN_THE_PAST_12_MONTHS_IN_2019_INFLATIONADJUSTED_DOLLARS.CSV', 2019)
INCOME_PAST_2017 = income_past_12_months('data/2017/INCOME_IN_THE_PAST_12_MONTHS_IN_2017_INFLATIONADJUSTED_DOLLARS.CSV', 2017)
INCOME_PAST_2015 = income_past_12_months('data/2015/INCOME_IN_THE_PAST_12_MONTHS_IN_2015_INFLATIONADJUSTED_DOLLARS.CSV', 2015)
INCOME_PAST_2014 = income_past_12_months('data/2014/INCOME_IN_THE_PAST_12_MONTHS_IN_2014_INFLATIONADJUSTED_DOLLARS.CSV', 2014)
INCOME_PAST_2013 = income_past_12_months('data/2013/INCOME_IN_THE_PAST_12_MONTHS_IN_2013_INFLATIONADJUSTED_DOLLARS.CSV', 2013)
INCOME_PAST_2012 = income_past_12_months('data/2012/INCOME_IN_THE_PAST_12_MONTHS_IN_2012_INFLATIONADJUSTED_DOLLARS.CSV', 2012)
INCOME_PAST_2011 = income_past_12_months('data/2011/INCOME_IN_THE_PAST_12_MONTHS_IN_2011_INFLATIONADJUSTED_DOLLARS.CSV', 2011)
INCOME_PAST_2010 = income_past_12_months('data/2010/INCOME_IN_THE_PAST_12_MONTHS_IN_2010_INFLATIONADJUSTED_DOLLARS.CSV', 2010)
```

```python
INCOME_PAST_12_MONTHS_STATUS = pd.concat([INCOME_PAST_2010, INCOME_PAST_2011, INCOME_PAST_2012, INCOME_PAST_2013, INCOME_PAST_2014, INCOME_PAST_2015, INCOME_PAST_2017, INCOME_PAST_2019, INCOME_PAST_2021, INCOME_PAST_2022])
INCOME_PAST_12_MONTHS_STATUS
```

```python
INCOME_PAST_12_MONTHS_STATUS.reset_index(inplace=True)
INCOME_PAST_12_MONTHS_STATUS.rename(columns={'index': 'Year'}, inplace=True)
INCOME_PAST_12_MONTHS_STATUS
```
